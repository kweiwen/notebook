```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal, fft, interpolate
import math
from cmath import exp, pi
from dtaidistance import dtw
from dtaidistance import dtw_visualisation as dtwvis
```

```python
# read raw data
def read_data(filename):
    result = []  
    with open(filename,'r') as file:  
        csvReader=csv.reader(file)  
        for line in csvReader:
            result.append(float(line[0]))
    return result
```

```python
dataSet1 = read_data("fft_data/FFT512complx_ADSP.dat")
dataSet2 = read_data("fft_data/FFT512complx_ADSP0407.dat")

s1 = np.array(dataSet1)
s2 = np.array(dataSet2)

distance, paths = dtw.warping_paths(s1, s2)
best_path = dtw.best_path(paths)
fig, axes = dtwvis.plot_warpingpaths(s1, s2, paths, best_path)
fig.set_size_inches(7,7)
```
