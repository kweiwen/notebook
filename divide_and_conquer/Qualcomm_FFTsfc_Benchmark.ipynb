```python
import numpy as np
import struct
import matplotlib.pyplot as plt
import IPython

def string_to_byte_array(input_data):
    output_data = list()
    output_data.append(int(input_data[0:2], base=16))
    output_data.append(int(input_data[2:4], base=16))
    output_data.append(int(input_data[4:6], base=16))
    output_data.append(int(input_data[6:8], base=16))
    return output_data

def byte_array_to_float_big(input_data):
    return struct.unpack('<f', bytes(input_data))[0]

little_ending = "0000803a"
print(little_ending[0:2])
print(little_ending[2:4])
print(little_ending[4:6])
print(little_ending[6:8])
print(byte_array_to_float_big(string_to_byte_array(little_ending)))
```

```python
# raw_value = ""
# raw_value = ""
raw_value = ""
```

```python
print(len(raw_value))
# buffer = np.ndarray(shape=(4096, 8), dtype=float, order='F')
buffer = np.ndarray(shape=(4096, 1), dtype=float, order='F')
for i in range(4096):
    little_ending = raw_value[i*8:(i+1)*8]
    buffer[i] = byte_array_to_float_big(string_to_byte_array(little_ending))
```

```python
input_data_real = buffer[0:2048:2]
input_data_imag = buffer[1:2048:2]

output_data_real = buffer[2048:4096:2]
output_data_imag = buffer[2049:4096:2]
output = output_data_real + output_data_imag*1j
```

```python
plt.title("Time Domain Data")
plt.plot(input_data_real)
plt.plot(input_data_imag)
plt.show()

```

```python
reference = np.linspace(0,1,1024, endpoint=False)
fd = np.fft.fft(reference)

plt.title("Numpy FFT Amplitude Response")
plt.plot(np.abs(fd))
plt.show()

plt.title("Numpy FFT Phase Response")
plt.plot(np.angle(fd))
plt.show()
```
