```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
N = 64
n = np.arange(N)
k = np.arange(N).reshape(N,1)
Zn = np.exp(-2j * np.pi * k * n / N)
```

```python
plt.plot(Zn[1].real)
plt.plot(Zn[1].imag)
plt.show()
```

```python
xn = np.arange(N)
yn = np.zeros(N)

fig = plt.figure(figsize=(12, 16))
ax = fig.add_subplot(projection='3d')

# plot real axis
ax.stem(xn, Zn.real[1], yn, orientation='y')

# plot imaginary axis
(markers, stemlines, baseline) = ax.stem(xn, yn, Zn.imag[1], linefmt='--')
plt.setp(markers, markeredgecolor="green")

# ax.set_xlim3d(0, size//2)
ax.set_ylim3d(-max(abs(Zn.real[1])), max(abs(Zn.real[1])))
ax.set_zlim3d(-max(abs(Zn.imag[1])), max(abs(Zn.imag[1])))  
ax.set_xlabel('Time(sample)')
ax.set_ylabel('Real Vector')
ax.set_zlabel('Imaginary Vector')

# ax.view_init(elev = elevation, azim=angle)
plt.show()
```
