```python
from math import pi
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
```

```python
sr = 48000
fc = 4000

alpha = (np.tan(pi * fc / sr) - 1) / (np.tan(pi * fc / sr) + 1)
numerator = [alpha, 1]
denominator = [1, alpha] 

w, h = scipy.signal.freqz(np.add(numerator, denominator)/2, denominator)

amplitude = 20 * np.log10(abs(h) / 2)
angle = np.angle(h)

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

plt.title('frequency response of allpass filter')
plt.xlabel(r'normalized frequency (x$\pi$rad/sample)')

ax1.plot(w/max(w), amplitude, 'g')
ax1.set_ylabel('amplitude (dB)', color='g')
ax1.grid()

ax2.plot(w/max(w), angle, 'b--')
ax2.set_ylabel('phase (radians)', color='b')

plt.show()


w, h = scipy.signal.freqz(numerator, denominator)

amplitude = 20 * np.log10(abs(h) / 2)
angle = np.angle(h)

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

plt.title('frequency response of allpass filter')
plt.xlabel(r'normalized frequency (x$\pi$rad/sample)')

ax1.plot(w/max(w), amplitude, 'g')
ax1.set_ylabel('amplitude (dB)', color='g')
ax1.grid()

ax2.plot(w/max(w), angle, 'b--')
ax2.set_ylabel('phase (radians)', color='b')

plt.show()
```
