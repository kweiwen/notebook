```python
from scipy.signal import chirp, spectrogram
import numpy as np
import matplotlib.pyplot as plt
import wavio
```

```python
Time = 0.01161
Sample = 512
t = np.linspace(0, Time, Sample)
w = chirp(t, f0=40, f1=8000, t1=Time, method='logarithmic')

plt.plot(t, w)
plt.title("Linear Chirp, f(0)=6, f(10)=1")
plt.xlabel('t (sec)')
plt.show()
```

```python
buffer = np.zeros(Sample*2)
for i, elements in enumerate(w):
    buffer[i*2] = elements 
    buffer[i*2+1] = 0

np.savetxt("chirp.dat", [buffer], delimiter=',\n', fmt='%1.19f')
```

```python
sr = 44100
output = np.zeros((len(w), 2))
output[:, 0] = w
output[:, 1] = w
wavio.write("chirp.wav", output, sr, sampwidth=3)
```
