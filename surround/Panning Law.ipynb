```python
from math import pi
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import wavio
import IPython
from scipy.io.wavfile import write, read
import sys
```

```python
arr = np.linspace(0, 1, 128, True)
```

```python
x = arr
y = 1 - x

plt.plot(x)
plt.plot(y)
plt.show()
```

```python
x = np.sqrt(arr)
y = np.sqrt(1 - arr)

plt.plot(x)
plt.plot(y)
plt.show()
```

```python
x = np.sin(arr * np.pi / 2)
y = np.sin((1 - arr) * np.pi / 2)

plt.plot(x)
plt.plot(y)
plt.show()
```
