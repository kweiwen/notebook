```python
from math import pi
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
import wavio
import IPython
```

```python
fs = 48000
frameSize = 1024
f = 440.0

xn = (np.sin(2*np.pi*np.arange(frameSize)*f/fs))

plt.plot(xn)
plt.show()
```

```python
X = np.fft.fft(xn)
```

```python
plt.plot(np.abs(X))
plt.show()
```

```python
plt.plot(np.angle(X))
plt.show()
```

```python
xn_ = np.array([])
for i in range(64):
    xn_ = np.append(xn_, xn)

plt.plot(xn_)
plt.xlim(0, 2048)
plt.show()
```

```python
wavio.write("xn_.wav", xn_, fs, sampwidth=3)
IPython.display.display(IPython.display.Audio("xn_.wav", rate=fs))
```
