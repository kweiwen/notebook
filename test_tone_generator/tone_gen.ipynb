```python
from scipy.signal import chirp, spectrogram
import numpy as np
import matplotlib.pyplot as plt
import wavio
```

```python
volume = 0.5     # range [0.0, 1.0]
fs = 512         # sampling rate, Hz, must be integer
duration = 1.0   # in seconds, may be float
frameSize = 512  # frame Size of audio processing
f = 2.0          # sine frequency, Hz, may be float

# generate samples, note conversion to float32 array
w = (np.sin(2*np.pi*np.arange(frameSize)*f/fs))

x_axis = np.arange(len(w))
plt.plot(x_axis, w)
plt.title("Sine Tone")
plt.xlabel('t (sample)')
plt.show()
```

```python
def saveAsDat(fname, data):
    buffer = np.zeros(len(data)*2)
    for i, element in enumerate(data):
        buffer[i*2] = element
        buffer[i*2+1] = 0
    np.savetxt(fname, [buffer], delimiter=',\n', fmt='%1.19f')

def saveAsWav(fname, data, sampleRate):
    output = np.zeros((len(data), 2))
    output[:, 0] = data
    output[:, 1] = data
    wavio.write(fname, output, sampleRate, sampwidth=3)    
```

```python
saveAsDat("TEST.dat",w)
saveAsWav("TEST.wav",w, 44100)
```
