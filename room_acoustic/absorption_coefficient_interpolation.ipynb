```python
import numpy as np
from scipy.fftpack import dct
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
def rescale(x_tick, x_value, y_tick):
    interpolator = interp1d(
        np.log2(x_tick),
        x_value,
        fill_value="extrapolate",
        kind="quadratic",
    )

    y_value = interpolator(np.log2(y_tick))

    # now clip between 0. and 1.
    y_value[y_value < 0.0] = 0.0
    y_value[y_value > 1.0] = 1.0
    
    return y_value
```

```python
x_tick = [125, 250, 500, 1000, 2000, 4000, 8000]
x_value = [0.02, 0.02, 0.03, 0.03, 0.04, 0.05, 0.05]

y_tick = [125, 250, 500, 1000, 2000, 4000, 8000, 16000]
y_value = rescale(x_tick, x_value, y_tick)

plt.plot(x_tick, x_value, '-o', y_tick, y_value, '--')
plt.show()
```
